Bài 3: Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def is_prime(num):
    if num < 2:
        return False
    else:
        for i in range(2, int(num**0.5)+1):
            if num % i == 0:
                return False
    return True
n = int(input())
print(is_prime(n))
        
```

Bài 4: VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên

```python
def so_hoan_hao(num):
    tong = 0
    for i in range(1, int(num**0.5)+1):
        if num % i == 0:
            tong += i
            if tong <= num:
                return True
            
    return False
n = int(input())
print(so_hoan_hao(n))
```

Bài 6: Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!)

```python
def giai_thua(num):
    total = 1
    for i in range(1, num+1):
        total *= i
    return total
n = int(input())
print(giai_thua(n))
```

Bài 8: Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng
của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2.
Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này.
Khoảng cách Hamming là cái tên được đặt theo tên của Richard Hamming, người giới thiệu lý
thuyết này trong tài liệu có tính cơ sở của ông về mã phát hiện lỗi và sửa lỗi (error-detecting and
error-correcting codes). Nó được sử dụng trong kỹ thuật viễn thông để tính số lượng các bit trong
một từ nhị phân (binary word) bị đổi ngược, như một hình thức để ước tính số lỗi xảy ra trong quá
trình truyền thông, và vì thế, đôi khi, nó còn được gọi là khoảng cách tín hiệu (signal distance). Việc
phân tích trọng số Hamming của các bit còn được sử dụng trong một số ngành, bao gồm lý thuyết
tin học, lý thuyết mã hóa, và mật mã học. Tuy vậy, khi so sánh các dãy ký tự có chiều dài khác
nhau, hay các dãy ký tự có xu hướng không chỉ bị thay thế không thôi, mà còn bị ảnh hưởng bởi dữ
liệu bị lồng thêm vào, hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách
Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn.

```python
def khoang_cach(a, b):
    c = a ^ b
    distance = 0
    while c > 0:
        if c % 2 == 1:
            distance += 1
        c //= 2
    return distance
a, b = map(int, input().split())
print(khoang_cach(a,b))

        

    
    
```

Bài 10: Hai từ được gọi là đẳng cấu nếu các kí tự trong một từ có thể được ánh xạ để được từ còn lại. Ánh
xạ một kí tự nghĩa là thay đổi toàn bộ các kí tự đó trong từ bởi một kí tự khác. Thứ tự của các kí tự
được giữ không đổi. Hai kí tự khác nhau không thể ánh xạ đến cùng một kí tự, nhưng một kí tự có
thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành
'z', 'b' thành 'b' và 'c' thành 'x'.
Viết hàm trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai
từ đầu vào có cùng độ dài.

```python
def dang_cau(a, b):
    chuyen_a_sang_b = {}
    chuyen_b_sang_a = {}
    for i, k in zip(a, b):
        if i in chuyen_a_sang_b:
            if chuyen_a_sang_b[i] != k:
                return False
            else:
                chuyen_a_sang_b[i] = k
        if k in chuyen_b_sang_a:
            if chuyen_b_sang_a[k] != i:
                return False
            else:
                chuyen_b_sang_a[k] = i
    return True
a, b = map(str, input().split())
print(dang_cau(a, b))
```

```python
def dang_cau(a, b):
    if len(a) != len(b):
        return False
    return len(set(a)) == len(set(b)) == len(set(zip(a, b)))
        
a, b = map(str, input().split())
print(dang_cau(a, b))
```

0 <= |r| < |divisor|.

Bài 15: Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông.

```python
def is_right_triangle(a, b, c):
    if a > 0 and b > 0 and c > 0 and a + b > c and a + c > b and b + c > a:
        if a**2 + b**2 == c**2 or b**2 + c**2 == a**2 or a**2 + c**2 == b**2:
            return True
    return False
a, b, c = map(int, input().split())
print(is_right_triangle(a, b, c))
```
