Bài 1: Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
def is_Amstronhg(x):
    if x < 0:
        return False
    a = str(x)
    b = len(a)
    tong = sum(int(digit)**b for digit in a )
    if tong == x:
        return True
    return False
n = int(input())
print(is_Amstronhg(n))

    
    
```

Bài 8: Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0.


```python
def quadratic_roots_count(a, b, c):
    delta = b**2 - 4*a*c
    
    nghiem_dau = (-b + delta**0.5)/(2*a)
    
    nghiem_sau = (-b-delta**0.5)/(2*a)
    
    return nghiem_dau, nghiem_sau
a, b, c = map(int, input().split())
print(quadratic_roots_count(a, b, c))
```

```python
def are_collinear(x1, y1, x2, y2, x3, y3):
    a = (x2-x1)
    b = (y2-y1)
    c = (x3-x1)
    d = (y3-y1)
    if a * d == c * b:
        return True
    return False
x1, y1, x2, y2, x3, y3 = map(int, input().split())
print(are_collinear(x1, y1, x2, y2, x3, y3))
```

```python
def point_in_quadrilateral(quad, p):
   
    n = len(quad)
    positive = 0
    negative = 0
    
    for i in range(n):
        
        x1, y1 = quad[i]
        x2, y2 = quad[(i + 1) % n] 
        
        x, y = p
        
        
        d = (x2 - x1) * (y - y1) - (y2 - y1) * (x - x1)
        
        if d > 0:
            positive += 1
        elif d < 0:
            negative += 1
            
    
    return positive == 0 or negative == 0


quad = [(0,0), (4,0), (4,3), (0,3)]
p = (2,1)

if point_in_quadrilateral(quad, p):
    print(f"Điểm {p} nằm trong hoặc trên biên tứ giác.")
else:
    print(f"Điểm {p} nằm ngoài tứ giác.")
```

```python
def so_luong_so(danh_sach):
    so = 0
    for num in danh_sach:
        s = str(abs(num))
        for i in s:
            if i != 0:
                so += 1
        return so

n = list(map(int, input().split()))
print(so_luong_so(n))
    
```

Bài 13: Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).

```python
import math
def tinh_goc(x1, y1, x2, y2, x3, y3):
    a = x2-x1
    b = y2-y1
    c = x3-x1
    d = y3 - y1
    ab = (a**2 + b**2)**0.5
    cd = (c ** 2 + d **2)**0.5
    
    if ab == 0 or cd == 0:
        return 0
    goc = (a*c + b*d)/(ab*cd)
    goc = max(-1, min(1, goc))
    radian = math.acos(goc)
    degree = math.degrees(radian)
    return degree
    

x1, y1, x2, y2, x3, y3 = map(int, input().split())
print(tinh_goc(x1, y1, x2, y2, x3, y3))
```

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    if car_type == 4:
        opening_fare = 11000 
        price_per_km = 14500  
    else:
        opening_fare = 13000
        price_per_km = 16000

    
    if km <= 0.5:
        travel_fare = opening_fare
    else:
        travel_fare = opening_fare + (km - 0.5) * price_per_km

    
    waiting_fare = max(0, wait_minutes - 5) * 1000

    total_fare = travel_fare + waiting_fare

    
    if start_hour >= 22 or start_hour < 5:
        total_fare *= 1.1

    
    if km > 30:
        total_fare *= 0.95

    return round(total_fare)

res = smart_taxi_fare(4, 23, 35, 10)
print(f"Tổng cước taxi: {res:,} VNĐ")
```
